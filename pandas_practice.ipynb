```python
import pandas as pd

data = {
    "Name": ["Rahul", "Aman", "Neha", "Priya", "Rohit", "Sneha"],
    "Class": ["A", "A", "B", "B", "A", "B"],
    "Marks": [85, 78, 90, 60, 88, 72]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby("Class")["Marks"].mean()
```

```python
df.groupby("Class")["Marks"].sum()
```

```python
df.groupby("Class")["Name"].count()
```

```python
df.groupby("Class")["Marks"].agg(["mean", "max", "min"])
```

```python
df.groupby("Class")["Marks"].max()
```

Example Dataset (WITH Missing Values)

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Rahul", "Aman", "Neha", "Priya", "Rohit"],
    "Marks": [85, np.nan, 90, 60, np.nan],
    "Class": ["A", "A", "B", "B", "A"]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()

```

```python

df.fillna(method="ffill")
```

```python
df.groupby("Class")["Marks"].mean()

```
